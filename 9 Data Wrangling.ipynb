### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here

# Get the value counts for 'Country' and return a Series
df_country=df['Country'].value_counts()

# Sort the value counts alphabetically by index (country names)
df_country_sorted = df_country.sort_index(ascending=True)

# Set Pandas to display all rows
pd.set_option('display.max_rows', None)

# Print the sorted value counts
print(df_country_sorted)

# Count the number of rows
df_country.shape

# 184 rows (can use this to check modifications later on)

# From this I can see the same country written in multiple different ways
# Congo, Republic of the... and Democratic Republic of the Congo 
# Democratic People's Republic of Korea and Republic of Korea and South Korea
# Also, nomadic is not a country 


```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here

# From the Country analysis, I can see the same country written in multiple different ways
# Congo, Republic of the... and Democratic Republic of the Congo 
# Democratic People's Republic of Korea and Republic of Korea and South Korea

# Count number of unique rows before modifications
df['Country'].nunique()

# map Congo variations to 'Democratic Republic of the Congo'

df['Country'].replace('Congo, Republic of the...', 'Democratic Republic of the Congo', inplace=True)

# check this has worked

df['Country'].nunique()

# map Korea variations to 'South Korea'

df['Country'].replace('Democratic People\'s Republic of Korea', 'South Korea', inplace=True)
df['Country'].replace('Republic of Korea', 'South Korea', inplace=True)

df['Country'].nunique()

# Remove rows with nomadic as a country
# Drop rows where 'Country' column has the value 'Nomadic'
df_cleaned = df[df['Country'] != 'Nomadic']

# check this has worked

df_cleaned['Country'].nunique()
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here

## new function
def assign_experience_level(years):
    if years<=2:
        return 'Beginner'
    elif 3<= years <5:
        return 'Intermediate'
    else:
        return 'Advanced'

## remove rows with nan values
df_cleaned.dropna(subset=['YearsCodePro'], axis=0, inplace=True)

# change 'less than 1 year' to '0', 'more than 50 years' to '55'

df_cleaned['YearsCodePro'].replace({
    'Less than 1 year': '0',
    'More than 50 years': '55'
}, inplace=True)

## make sure data in YearsCodePro is all one type (int)
df_cleaned[['YearsCodePro']] = df_cleaned[['YearsCodePro']].astype('int')

## apply function to YearsCodePro column

df_cleaned['ExperienceLevel']=df_cleaned['YearsCodePro'].apply(assign_experience_level)

## print the updated dataframe

print(df_cleaned['ExperienceLevel'])
    
```
