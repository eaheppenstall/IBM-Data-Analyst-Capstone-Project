## Extract information from the given web site
You will extract the data from the below web site: <br> 


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Your code here
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

Download the webpage at the url


```python
#your code goes here
data=requests.get(url).text
```

Create a soup object


```python
#your code goes here
soup=BeautifulSoup(data, "html.parser")
```

Scrape the `Language name` and `annual average salary`.


```python
#your code goes here

import pandas as pd

# Create an empty DataFrame with the appropriate column names
language_salary = pd.DataFrame(columns=["Language name", "Annual Average Salary"])

# Assuming `soup` is already defined as your BeautifulSoup object
table = soup.find('table')

# Initialize a list to hold row data
rows_data = []

# Loop through the rows of the table and extract data
for row in table.find_all('tr'):
    cols = row.find_all('td')  # Get all columns (td elements)
    
    # Ensure there are enough columns to avoid IndexError
   
    language_name = cols[1].getText()  # Extract the language name
    annual_avg_salary = cols[3].getText()  # Extract the annual average salary
        
        # Append the row data to the list
    
    rows_data.append([language_name, annual_avg_salary])

# Convert the list of rows into a DataFrame
language_salary = pd.DataFrame(rows_data, columns=["Language name", "Annual Average Salary"])

# Show the first few rows of the resulting DataFrame
print(language_salary.head(11))
                                                         
```

Save the scrapped data into a file named *popular-languages.csv*


```python
# your code goes here


language_salary.to_csv('popular-languages.csv', index=False)

data = {'Language': ['Python', 'Java', 'R', 'Javascript', 'Swift', 'C++', 'C#', 'PHP', 'SQL', 'Go'],
       'Average Annual Salary': ['114383', '101013', '92037', '110981', '130801', '113865', '88726', '84727', '84793', '94082']}
df_new = pd.DataFrame(data)

df_new.head(10)

df_new['Average Annual Salary'] = df_new['Average Annual Salary'].astype(float)

df_new = df_new.set_index(df_new.columns[0])

df_sorted = df_new.sort_values(by='Average Annual Salary', ascending=False)

import matplotlib as mpl
import matplotlib.pyplot as plt

df_sorted.plot(kind='bar')
plt.title('Popular programming languages and their Average Annual Salary')
plt.ylabel('Average Annual Salary, $')
plt.xlabel('Language')
plt.show()






```
