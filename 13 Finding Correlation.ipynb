##### Step 1: Install and Import Required Libraries


```python
# Install the necessary libraries
!pip install pandas
!pip install matplotlib
!pip install seaborn

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
## Write your code here

!pip install numpy
import numpy as np

## Remove and nan values / make sure values are numeric in ConvertedCompYearly

df.dropna(subset=['ConvertedCompYearly'], axis=0, inplace=True)
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

#distribution is a line graph

df['ConvertedCompYearly'].plot(kind='line')
plt.title('Distribution of Converted Comp Yearly')
plt.show()

#histogram

count, bin_edges = np.histogram(df['ConvertedCompYearly'], bins=15)
df['ConvertedCompYearly'].plot(kind='hist', bins=bin_edges, edgecolor='black', alpha=0.7)
plt.title('histogram of Converted Comp Yearly')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('Frequency')
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
## Write your code here

df_groupone = df[['Employment', 'ConvertedCompYearly']]

# Filter the data for 'Employed, full-time' status
df_full_time = df_groupone[df_groupone['Employment'] == 'Employed, full-time']

# Calculate the median of 'ConvertedCompYearly' for the filtered data
median_compensation = df_full_time['ConvertedCompYearly'].median()

print("Median compensation for 'Employed, full-time':", median_compensation)

```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
## Write your code here

df_Country_vs_ConvertedCompYearly = df[['Country', 'ConvertedCompYearly']]

# Filter the data for Canada
df_Canada = df_Country_vs_ConvertedCompYearly[df_Country_vs_ConvertedCompYearly['Country'] == 'Canada']

df_Canada.plot(kind='box')
plt.title('Compensation in Canada')

# Highlight the median value
median = df_Canada['ConvertedCompYearly'].median()

# Add a horizontal line at the median value
plt.axhline(median, color='r', linestyle='--', label=f'Median: {median}')

# Add a text label for the median
plt.text(0, median + 5000, f'Median: {median}', horizontalalignment='center', color='red', fontsize=12)

plt.show()


# Filter the data for France
df_France = df_Country_vs_ConvertedCompYearly[df_Country_vs_ConvertedCompYearly['Country'] == 'France']

df_France.plot(kind='box')
plt.title('Compensation in France')

# Highlight the median value
median = df_France['ConvertedCompYearly'].median()

# Add a horizontal line at the median value
plt.axhline(median, color='r', linestyle='--', label=f'Median: {median}')

# Add a text label for the median
plt.text(0, median + 5000, f'Median: {median}', horizontalalignment='center', color='red', fontsize=12)

plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
## Write your code here

# Same task as in lab 12

# Calculate the mean and standard deviation
mean_ConvertedCompYearly = df['ConvertedCompYearly'].mean()
std_ConvertedCompYearly = df['ConvertedCompYearly'].std()

# Define the threshold (mean + 3 * std)
threshold = mean_ConvertedCompYearly + 3 * std_ConvertedCompYearly

# create new dataframe
df_cleaned = df[df['ConvertedCompYearly'] <= threshold]
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
## Write your code here
## Heatmap between ConvertedCompYearly and JobSatPoints_1

# Calculate the correlation matrix between the two variables
correlation_matrix = df_cleaned[['ConvertedCompYearly', 'JobSatPoints_1']].corr()

# Plot the heatmap of the correlation matrix
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, cbar_kws={'label': 'Correlation'})
plt.title('Correlation Heatmap of ConvertedCompYearly vs JobSatPoints_1')
plt.show()
```

<h3>Step 8: Scatter Plot for Correlations</h3>


**Task**: Create scatter plots to examine specific correlations between `ConvertedCompYearly` and `WorkExp`, as well as between `ConvertedCompYearly` and `JobSatPoints_1`.


```python
## Write your code here

## Scatter plot between ConvertedCompYearly and JobSatPoints_1

df_cleaned.plot(kind='scatter', x='ConvertedCompYearly', y='JobSatPoints_1', title='Correlation between Compensation and Job Satisfaction')
plt.ylabel('Job Satisfaction Points')
plt.xlabel('ConvertedCompYearly')
plt.show()


```
