### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here

!pip install matplotlib-venn

import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Define the sets
set_languages_have_worked_with = set(df['LanguageHaveWorkedWith'].dropna())
set_languages_want_to_work_with = set(df['LanguageWantToWorkWith'].dropna())

# Create the Venn diagram
plt.figure(figsize=(6, 6))
venn2([set_languages_have_worked_with, set_languages_want_to_work_with], 
      set_labels=('Languages Have Worked With', 'Languages Want To Work With'))

# Add a title
plt.title("Overlap of Languages Worked With and Languages Want To Work With")

# Display the plot
plt.show()

```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here

#scatter graph to look at correlation between two variables

import matplotlib.pyplot as plt
import pandas as pd

# Drop rows with NaN values in the specified columns
df.dropna(subset=['YearsCodePro', 'JobSat'], axis=0, inplace=True)

# Replace 'Less than 1 year' with '0' and 'More than 50 years' with '55'
df['YearsCodePro'].replace({'Less than 1 year': '0', 'More than 50 years': '55'}, inplace=True)

# Convert 'YearsCodePro' and 'JobSat' to numeric values
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce') 
df['JobSat'] = pd.to_numeric(df['JobSat'], errors='coerce')

# Drop rows again if there are any NaN values left after conversion
df.dropna(subset=['YearsCodePro', 'JobSat'], axis=0, inplace=True)

# Create the scatter plot
df.plot(kind='scatter', x='JobSat', y='YearsCodePro')

# Customize plot labels and title
plt.title('Correlation between Job Satisfaction and Years of Coding Experience')
plt.ylabel('Experience')
plt.xlabel('Job Satisfaction')

# Show the plot
plt.show()

```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here

crosstab_result = pd.crosstab(df['Employment'], df['EdLevel'])
print(crosstab_result)

## stacked bar chart

crosstab_result.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title('Stacked Bar Chart: Employment vs Education Level')
plt.xlabel('Employment Type')
plt.ylabel('Count of Education Level')
plt.legend(title='Education Level', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
